```python
# --- Phase 2: Data Preparation ---

# Step 1: Separate features (X) and target (y)
X = df.drop('current_value', axis=1) # X is everything EXCEPT the 'current_value' column
y = df['current_value']             # y IS ONLY the 'current_value' column

# Display the first 5 rows of each to confirm the split was successful
print("--- Features (X) ---")
print(X.head())
print("\n" + "="*30 + "\n")
print("--- Target (y) ---")
print(y.head())
```

```python
import pandas as pd

# Step 2: Apply One-Hot Encoding to categorical columns
X_encoded = pd.get_dummies(X, columns=['team', 'position'])

# Display the result to see the transformation
print(f"Shape of X before encoding: {X.shape}")
print(f"Shape of X after encoding: {X_encoded.shape}")

print("\n--- Encoded Features (first 5 rows) ---")
print("Notice the original 'team' and 'position' columns are replaced by many new 0/1 columns.")
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split

# Step 3: Split the data into training and testing sets
# X_encoded is our full set of numerical features
# y is our target (current_value)
# test_size=0.2 means 20% of the data will be for testing
# random_state=42 ensures we get the same "random" split every time we run this code
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# Print the shapes of the new datasets to confirm the split
print(f"Original number of players: {len(df)}")
print("-" * 30)
print(f"Number of players in the Training set: {len(X_train)}")
print(f"Number of players in the Testing set: {len(X_test)}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np # We need numpy to calculate the square root

# --- Phase 3: Model Training (Corrected) ---

# This part is the same as before
rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
print("Training the Random Forest model... (This may take a moment)")
rf_model.fit(X_train, y_train)
print("Training complete!")

# --- THE FIX IS HERE ---
# Make predictions on the training data
training_predictions = rf_model.predict(X_train)

# 1. Calculate the Mean Squared Error first
training_mse = mean_squared_error(y_train, training_predictions)
# 2. Then, take the square root to get the RMSE
training_rmse = np.sqrt(training_mse)
# --- END OF FIX ---

print("-" * 30)
print(f"Model's average error on the training data (RMSE): {training_rmse:,.2f}")
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# --- Step 2: Evaluate the model on the unseen test data ---

print("Making predictions on the test set...")
test_predictions = rf_model.predict(X_test)

# Calculate the Root Mean Squared Error (RMSE) on the test set
test_mse = mean_squared_error(y_test, test_predictions)
test_rmse = np.sqrt(test_mse)

print("-" * 30)
print(f"Training RMSE: {training_rmse:,.2f}")
print(f"Test RMSE:     {test_rmse:,.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Get the feature importances from our trained model
importances = rf_model.feature_importances_

# Create a pandas Series for easier viewing, sorted by importance
feature_importances = pd.Series(importances, index=X_encoded.columns).sort_values(ascending=False)

# Plot the top 20 most important features
plt.figure(figsize=(12, 10))
sns.barplot(x=feature_importances.head(20), y=feature_importances.head(20).index)

# Add labels and title
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.title("Top 20 Most Important Features")
plt.show()
```

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# --- Phase 4: Model Improvement ---
# Step 1: Train a simpler model by tuning a hyperparameter (max_depth)

print("Training a new, simpler model with max_depth=10...")
rf_tuned_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=10,         # This is the new setting to limit tree complexity
    random_state=42,
    n_jobs=-1
)
rf_tuned_model.fit(X_train, y_train)
print("Training complete!")

# --- Evaluate the new model's performance ---
# On training data
tuned_train_preds = rf_tuned_model.predict(X_train)
tuned_train_rmse = np.sqrt(mean_squared_error(y_train, tuned_train_preds))

# On test data
tuned_test_preds = rf_tuned_model.predict(X_test)
tuned_test_rmse = np.sqrt(mean_squared_error(y_test, tuned_test_preds))

# --- Compare the results ---
print("\n" + "="*40)
print("--- PERFORMANCE COMPARISON ---")
print(f"Original Model -> Train RMSE: {training_rmse:,.2f} | Test RMSE: {test_rmse:,.2f}")
print(f"Tuned Model    -> Train RMSE: {tuned_train_rmse:,.2f} | Test RMSE: {tuned_test_rmse:,.2f}")
print("="*40)
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.metrics import mean_squared_error

# --- Step 2: Automated Tuning with GridSearchCV ---

# 1. Define the grid of settings we want to test
param_grid = {
    'max_depth': [15, 20, 25],       # Let's try deeper trees
    'min_samples_split': [2, 5, 10], # The minimum number of samples required to split a node
    'n_estimators': [100]            # We'll keep the number of trees the same
}

# 2. Create the Grid Search tool
# Note: scoring uses 'neg_mean_squared_error' because the tool tries to find the highest score.
# By making the error negative, the highest score is the one closest to zero.
grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42, n_jobs=-1),
                           param_grid=param_grid,
                           scoring='neg_mean_squared_error',
                           cv=3, # Use 3-fold cross-validation
                           verbose=2) # This will print progress updates

# 3. Run the search on your training data
print("Running Grid Search to find the best model settings... This will take a while!")
grid_search.fit(X_train, y_train)

# 4. Get the best model found by the search
best_rf_model = grid_search.best_estimator_
print("\nBest settings found: ", grid_search.best_params_)

# 5. Evaluate this new best model on the test set
best_model_predictions = best_rf_model.predict(X_test)
best_model_rmse = np.sqrt(mean_squared_error(y_test, best_model_predictions))

# --- Final Comparison ---
print("\n" + "="*40)
print("--- FINAL PERFORMANCE ---")
print(f"Original Model Test RMSE: {test_rmse:,.2f}")
print(f"Best Model (from Grid Search) Test RMSE: {best_model_rmse:,.2f}")
print("="*40)
```
